```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Imagenette
categories = ('tench', 'English springer', 'cassette player', 'chain saw', 'church', 'French horn', 'garbage truck',
              'gas pump', 'golf ball', 'parachute')
counts = {
    'train': (963, 955, 993, 858, 941, 956, 961, 931, 951, 960),
    'validation': (387, 395, 357, 386, 409, 394, 389, 419, 399, 390)
}


x = np.arange(len(categories))
width = 0.3
multiplier = 0.5

fig, ax = plt.subplots(layout='constrained', figsize=(10, 5))

for type, count in counts.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, count, width, label=type)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel("Count")
ax.set_title("Number of Images Contained in Training and Validation Set in Imagenette", fontsize=16)
ax.set_xticks(x + width, categories, rotation=45)
ax.legend(loc='upper left')
ax.set_ylim(0, 1200)
plt.savefig('Imagenette.png')
plt.show()
```
